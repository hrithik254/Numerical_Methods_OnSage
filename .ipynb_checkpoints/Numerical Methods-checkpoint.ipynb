```python
var('x a n')
a = 0
error_list = []
rel_error_list = []
trunc_value_list = []
true_value = sin(pi/2)
approx_value = 1 
error = 0
t = (pi/2).n() #to get numerical value of pi/2
for i in range(8):
    trunc_value_list.append(i+1)
    
    f(x) = taylor(sin(x),x,0,i+1) #range function goes from 0 to 9
    
    realtive_error = error/f(t)
    rel_error_list.append(realtive_error)
    
    approx_value = f(t)
    error = true_value-approx_value
    error_list.append(error)
rel_error_list = rel_error_list[1:]
points = list(zip(trunc_value_list,error_list)) #to create a list of x,y coordinates where x being trunc_value y being error
list_plot(points, plotjoined=True, title= 'truncation error in Sin(pi/2) evaluated upto 8 terms of taylor series')
```

```python
def error_list_generator(function,variable,point,a_value,terms):
    
    var('variable')
    x = variable
    a = a_value.n()
    point = point.n()
    
    error_list = []
    rel_error_list = []
    trunc_value_list = []
    
    true_value = function(point)
    
    approx_value = 1 
    error = 0

    for i in range(terms):
        trunc_value_list.append(i+1)

        g(x) = taylor(function(x),x,a,i+1) #range function goes from 0 to 9

        realtive_error = error/g(point)
        rel_error_list.append(realtive_error)

        approx_value = g(point)
        error = abs(true_value-approx_value)
        error_list.append(error)
    rel_error_list = rel_error_list[1:]
    return trunc_value_list, error_list, rel_error_list, true_value

error_list_generator(sin,x,pi/2,0,8)
```

```python
t = error_list_generator(cos,x,pi,0,11) #t is a tuple
t[1]
error_points = list(zip(t[0],t[1]))
rel_error_points = list(zip(t[0],t[2]))
true_value = t[3]

list_plot(error_points,plotjoined=True, title = 'Truncation error in cos(pi) evaluated upto 11 terms of taylor series')
```

The in the taylor series expansions of the sine or cosine function there are only off and even terms so at regular intervals of one term the term of the expansion is zero and hence we see same error in two consecutive values
