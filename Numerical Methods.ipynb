```python
def error_list_generator(function,variable,point,a_value,terms):
    
    var('variable')
    x = variable
    a = a_value.n()
    point = point.n()
    
    error_list = []
    rel_error_list = []
    trunc_value_list = []
    
    true_value = function(point)
    
    approx_value = 1 
    error = 0

    for i in range(terms):
        trunc_value_list.append(i+1)

        g(x) = taylor(f(x),x,a,i+1) #range function goes from 0 to 9

        realtive_error = error/g(point)
        rel_error_list.append(realtive_error)

        approx_value = g(point)
        error = abs(true_value-approx_value)
        error_list.append(error)
    rel_error_list = rel_error_list[1:]
    return trunc_value_list, error_list, rel_error_list, true_value

error_list_generator(sin,x,pi/2,0,8)
```
